## DATA 512 Final Project Report: Analysis of World University Rankings

### 1. Introduction
Over the past years, university rankings have become one of essential references for applicants before submitting their applications. Knowing which countries have most globally ranked universities can give people who want to study abroad a clear direction on choosing countries. We also perform analysis to find reason why these countries have more globally ranked universities than others. Generally, ranking systems evaluate universities in many different aspects, including citations, research, influences and etc. Applicants may want to know which indicators can provide them more valuable information besides total score. While there are many different existing ranking systems, we used data from one of the most authoritative ranking systems, the Times Higher Education World University Ranking, as primary dataset on most analysis. Meanwhile, we used data from the other influential ranking system to make comparisons.

### 2. Background
More than 4.5 million students are enrolled in university-level education outside their country of origin and this number is continually growing. Over the past years, as a new form of higher education evaluation system, university ranking has become prevalence all over the world. World university rankings are widely used to measure the competitiveness of instructing institutions and regarded as one of crucial factors to consider for applicants of overseas universities. From the perspective of schools, it is important to know which objective indicators have the largest impact on their rankings so that they can improve practices and ranks to raise their reputations. On the other hand, it is also important for the people who are seeking for universities abroad to know which countries have most high-quality universities so that they can have a clearer direction on their applications. 

As a human-centered based project, I start the exploration of world university rankings with several research questions and corresponding hypotheses:
##### RQ1: What factors contribute to world university rankings a lot?
The rankings are calculated based on various measures. In this research question, I want to figure out what factors really make an educational institution ranks higher than the others. 
##### H1a: Universities with higher citation scores are more likely to rank higher.
Unlike national ranking systems, global university rankings may focus more on the citation area, other factors such as teaching and learning environment have less impact on the final ranking results.
##### H1b: Student-staff ratio affects rankings to a great extent. Higher ratio leads to lower ranking.
Student-staff ratio means number of students per staff. Students may get more attention in the universities with lower ratios which leads to higher education quality.
##### RQ2: Which countries have the most high-ranked universities on average? Why these countries have more top universities?
As far as I know, lots of globally-ranked universities are in the US. But how about other countries? In this research question, I would like to find out 10 highest-ranked countries in terms of number of top globally-ranked universities and what these countries have in common.
##### H2a: Countries have most high-ranked universities are wealthy developed countries. 
The governments of wealthy developed countries can invest more money in education. As a result, developed countries have more top universities than developing countries.
##### H2b: The population of a country affects the quality of its education. Populous countries might have more globally-ranked universities than the ones with small population. 
Populous countries have higher demand of educational resources. The larger the number of universities in a country, the greater the chances of these universities are ranked in top global universities.
##### RQ3: Do any biases might exist in ranking systems from various countries?
Different ranking systems have different preferences of indicator selections and even their evaluation standards are quite different. Therefore, there might be some biases between these ranking systems. In this research question, I want to test bias cause by regional difference.
##### H3a: A ranking system might include more globally-ranked universities of its producer’s original country than the system produced by another country.

Importing Data

```python
import csv
from itertools import islice
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
# get data from Center for World University Ranking
cwur = []
with open('cwurData.csv', newline='') as cwurData:
    readCSV = csv.reader(cwurData,delimiter=',')
    for row in readCSV:
        cwur.append(row)
# get data from Times Higher Education World University Ranking
times = []
with open('timesData.csv', newline='') as timesData:
    readCSV = csv.reader(timesData,delimiter=',')
    for row in readCSV:
        times.append(row)
# get data from population data of a variety of countries in 2015
population = []
with open('Population Mid-2015.csv') as population_file:
    reader = csv.reader(population_file)
    # note that first row is title; the second and last two rows are blank
    # skip first and last two rows in the csv file
    for row in islice(reader,2,213):
        population.append([row[0],row[4]])
```

Extract top 200 universities in 2015 from the THE ranking dataset

```python
# Convert THE ranking data to dataframe
timesDf = pd.DataFrame(times[1:],columns = times[0])
# Keep data from 2015
times2015 = timesDf[timesDf['year'] == '2015']
# Keep top 200 globally ranked universities in 2015
top200_idx_the = times2015[times2015['world_rank'] == '200'].index.values.tolist()[-1]
times2015_top200 = times2015.loc[:top200_idx_the,:]
```

List 10 example data of top 200 globally ranked universities in 2015

```python
times2015_top200.head(10)
```

Extract columns related to citation scores

```python
times_citation = times2015_top200[['world_rank','university_name','country','citations','total_score']]
```

##### RQ1: What factors contribute to world university rankings a lot?
###### H1a: Universities with higher citation scores are more likely to rank higher.

Plot citation scores and total score of top 200 universities

```python
%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib
# Plot citation score and total score for top 200 universities
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(times_citation['world_rank'].astype(int),times_citation['citations'].astype(float),label="citation_score",marker='o')
ax.plot(times_citation['world_rank'].astype(int),times_citation['total_score'].astype(float),label="total_score")
plt.title('Global Ranked Top 200 Universities Citation Scores VS Total Scores')
plt.xlabel('World Rank For University')
plt.ylabel('Score')
# Add a legend
ax.legend(loc='lower left', shadow=True)
```

##### Findings:
The overall observation is that the higher the ranking, the greater the citation score. Schools rank high tend to have smaller variation in citation scores. On the other hand, schools rank low have bigger variations in citation scores. Some of their scores are good and in parity with high ranking schools, while other scores are really low. It is likely that some of the research labs in the low ranking schools are strong and other labs are weak. For high ranking schools, most of their research labs are strong and produce high impacting, high citation publications.

##### H1b: Student-staff ratio affects rankings to a great extent. Higher ratio leads to lower ranking.

Extract columns related to student-staff-ratio index

```python
times_studentStaffRatio = times2015_top200[['world_rank','university_name','country','student_staff_ratio','total_score']]
```

```python
import numpy as np
# Replace blanks with NaN
times_studentStaffRatio['student_staff_ratio'] = times_studentStaffRatio['student_staff_ratio'].replace(r'\s+',np.nan,regex=True).replace('',np.nan)
times_studentStaffRatio['student_staff_ratio'] = np.array(times_studentStaffRatio['student_staff_ratio'])
```

```python
# Convert string to float if the value is not NaN
studentStaffRatio = []
for x in times_studentStaffRatio['student_staff_ratio']:
    if (pd.isnull(x)):
        studentStaffRatio.append(x)
        continue
    else:
        x = float(x)
        studentStaffRatio.append(x)
```

```python
times_studentStaffRatio['student_staff_ratio'] = studentStaffRatio
```

Plot student and staff ratio and total score of top 200 universities

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(times_studentStaffRatio['world_rank'].astype(int),times_studentStaffRatio['student_staff_ratio'],label="student_staff_ratio",marker='o')
ax.plot(times_studentStaffRatio['world_rank'].astype(int),times_studentStaffRatio['total_score'].astype(float),label="total_score")
plt.title('Global Ranked Top 200 Universities Student and Staff Ratio VS Total Scores')
plt.xlabel('World Rank For University')
plt.ylabel('Score for University or Index for Student and Staff Ratio')
# Add a legend
ax.legend(loc='upper right', shadow=True)
```

##### Findings:
It is observed that student staff ratio is low in high ranking universities. It is likely a way to keep their education quality high. Interesting, mid and low ranking universities can have big variance on student staff ratio. Both high and low student staff ratio can occur in mid and low ranking universities. Student staff ration does not appear to be a linear factor that impact university ranking score.

#### RQ2: Which countries have the most high-ranked universities on average? Why these countries have more top universities?

```python
# Extract column "country" from top 200 global ranked universities in 2015
times2015_top200_country = times2015_top200.iloc[:,2].tolist()
# Find 10 highest-ranked countries have most number of top globally-ranked universities
from collections import Counter
count_country = Counter(times2015_top200_country)
top10_countries = count_country.most_common(10)
```

#### H2a: Countries have most high-ranked universities are wealthy developed countries. 

```python
# Plot bar chart to visualize 10 highest-ranked countries from THE rankings
plt.subplots(figsize=(20,10))
plt.bar(range(len(top10_countries)), [val[1] for val in top10_countries], align='center')
plt.xticks(range(len(top10_countries)), [val[0] for val in top10_countries])
plt.xticks(rotation=70)
for x, y in zip(range(len(top10_countries)), [val[1] for val in top10_countries]):
    plt.annotate(str(y),xy=(x, y + 1.1))
plt.title("10 highest-ranked countries in terms of number of top globally-ranked universities(THE Data)")
plt.ylabel("Number of Universities")
plt.xlabel("Country Name")
plt.show()
```

##### Findings:
The top 10 counties with most high-ranking universities are all developed countries. Developed countries has a better education system, which may in turn contribute to their development. U.S. and U.K. rank number one and number two in this 10 countries. The universities from U.S. and U.K. combined account for more than half of the top 200 universities. This is consistent with the status of the two countries. Both U.S. and U.k. have a lot of high-tech companies, which is likely a result of talent trained from those universities.

#### H2b: The population of a country affects the quality of its education. Populous countries might have more globally-ranked universities than the ones with small population. 

```python
# Convert data (country and population) from the population file to dataframe
population_data = pd.DataFrame(population[1:],columns=population[0])
# Renames the columns with suitable names
population_data.columns = ["Location","Population"]
```

```python
top10_countries_population = []
for country in [val[0] for val in top10_countries]:
    if country == "United States of America":
        p = int(population_data[population_data["Location"] == "United States"].iloc[0][1].replace(",",""))
        top10_countries_population.append(p)
    else:
        p = int(population_data[population_data["Location"] == country].iloc[0][1].replace(",",""))
        top10_countries_population.append(p)
```

```python
fig, ax1 = plt.subplots(figsize=(20,10))
ax2 = ax1.twinx()
ax1.bar(range(len(top10_countries)), [val[1] for val in top10_countries],width = -0.2,align = "edge",label = "Number of Universities")
plt.xticks(range(len(top10_countries)), [val[0] for val in top10_countries])
plt.xticks(rotation=70)
ax2.bar(range(len(top10_countries)), top10_countries_population,width = 0.2,color = "red",align = "edge",label = "Population")
ax1.set_ylabel('Number of Universities')
ax2.set_ylabel('Population')
ax1.set_xlabel('Country Name')
plt.title("10 highest-ranked countries in terms of number of top globally-ranked universities VS their population")
ax1.legend(loc='upper center')
ax2.legend(loc='upper right')
```

##### Findings:
U.S. has most of the population in those 10 countries as well as most of the high ranking universities. Japan has the second largest population in the 10 countries. However, the number of high ranking universities in Japan is not the second place. There can be multiple factors why Japan does not have many high ranking universities. One of the factor might the duration of Japan’s current education system is not as long as U.S. or U.K. China and India are counties with high populations, however, they are not on the list. It is likely their universities systems need more time to develop and catch up.

##### RQ3: Do any biases might exist in ranking systems from various countries?

#### H3a: A ranking system might include more globally-ranked universities of its producer’s original country than the system produced by another country.

```python
# Convert CWUR ranking data to dataframe
cwurDf = pd.DataFrame(cwur[1:],columns = cwur[0])
# Keep data from 2015
cwur2015 = cwurDf[cwurDf['year'] == '2015']
# Keep top 200 globally ranked universities in 2015
top200_idx_cwur = cwur2015[cwur2015['world_rank'] == '200'].index.values.tolist()[-1]
cwur2015_top200 = cwur2015.loc[:top200_idx_cwur,:]
# Extract column "country" from top 200 global ranked universities in 2015
cwur2015_top200_country = cwur2015_top200.iloc[:,2].tolist()
# Find 10 highest-ranked countries have most number of top globally-ranked universities
from collections import Counter
count_country = Counter(cwur2015_top200_country)
top10_countries_cwur = count_country.most_common(10)
```

```python
# Plot bar chart to visualize 10 highest-ranked countries from CWUR rankings
plt.subplots(figsize=(20,10))
plt.bar(range(len(top10_countries_cwur)), [val[1] for val in top10_countries_cwur], align='center')
plt.xticks(range(len(top10_countries_cwur)), [val[0] for val in top10_countries_cwur])
plt.xticks(rotation=70)
for x, y in zip(range(len(top10_countries)), [val[1] for val in top10_countries_cwur]):
    plt.annotate(str(y),xy=(x, y + 1.1))
plt.title("10 highest-ranked countries in terms of number of top globally-ranked universities(CWUR Data)")
plt.ylabel("Number of Universities")
plt.xlabel("Country Name")
plt.show()
```

##### Findings:
Here a different ranking dataset is used to compare with the previous dataset. The previous dataset produced by U.K. has more U.K. universities listed. The dataset from the Center for World University Rankings produced by Saudi Arabia has less U.K. universities. It is mostly likely the U.K. based dataset has a bias towards U.K. universities. On the other hand, the Saudi Arabia based ranking has more Asian university such as South Korean.

### 5. Discussion
Overall, I believe the project and its outcomes are very helpful in the related studies of university rankings. Future study directions on existing bias in different ranking systems are also suggested. 

One interesting finding is that outstanding universities all put a lot of emphasize on research, and as a result have high citation scores. This is consistent with how those universities select professors. To become tenure, professors at those universities need to work hard on research and provide or guide their students to provide high quality publications. Outstanding universities tend to attract outstanding professors with their reputation, salaries, etc. And outstanding professors provide excellent research and publication. Those outstanding universities usually have sufficient funding to gather quite a few professors, and thus build up strong research ability. One fact is that universities with relative low ranking still can have high citation scores. It is not uncommon that some labs are outstanding at certain universities because some professors. For example, some professors may like to stay at Universities of Florida because of the weather of Florida.

Student staff ratio does not appear to have a direct relation with respect to university ranking. Some may argue that with more faculty per student, student may receive more instruction and better development. That might be true for elementary and middle school. In elementary and middle school, students need to development their mental and the ways of instruction highly rely on instructor. Student mostly passively receive knowledge from instructors. For university student, most of the time, one need to develop research ability and pick up the topic by oneself. The reach and richness of instruction material is largely enhanced by course website and Internet. A student will be able to be exposed to sufficient instructional materials. A classroom of small size versus a classroom of large size may not have significant impact on students’ mastery of course material.

Top 10 countries with most high ranking universities are all developed countries. This result makes a lot of sense. Advanced universities are more likely to produce breakthrough in research and advancement in technology. The advancement in technology will in turn help countries develop faster. More developed countries have more economic power to fund university research. This provide a positive feedback loop. We can take a look at how Internet is developed. It was some Berkeley research project and gradually adopted by people. The development of Internet boosted a lot of economic growth in U.S. and excellent companies. On the other hand, underdeveloped countries usually haven’t realized the importance of world class university or do not have the funding to develop such university.

One interesting finding is that population does not appear to be a factor of whether a country will have more high ranking universities. U.S. has a large population and most advanced universities in the world. Japan, with the second largest population, only has less than 10 universities rank in the top 200. Other countries rank the top 10 have a medium or small population. India and China, the two countries with most population in the world, does not appear in the top 10 list. It is not difficult to see that development status rather than population impact whether a country will have more advanced high ranking universities.
